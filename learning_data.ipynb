```python
import pandas as pd

# Step 1 — Marks pehle alag list mein rakho
# Taaki result mein use kar sakein
marks = [60, 89, 55, 76, 90]

# Step 2 — DataFrame banao
# Har key = ek column
# Har value = us column ki list (sab same length!)
df1 = pd.DataFrame({
    "Name":    ["Karna", "Arjun", "Bheema", "Nakul", "Sahadev"],
    "Subject": ["Math", "Science", "English", "Math", "Science"],
    "Marks":   marks,
    # if/else pehle — for baad mein — yeh rule yaad rakho!
    "Result":  ["Pass" if i > 60 else "Fail" for i in marks]
})

print("Poora DataFrame:")
print(df1)
print()

# Step 3 — Sirf Pass students
pass_df = df1[df1["Result"] == "Pass"]
print("Pass students:")
print(pass_df)
print()

# Step 4 — 80 se zyada marks
above_80 = df1[df1["Marks"] > 80]
print("80 se zyada marks wale:")
print(above_80)
print()

# Step 5 — Har subject ka average
avg_marks = df1.groupby("Subject")["Marks"].mean()
print("Har subject ka average:")
print(avg_marks)

```

```python
import pandas as pd
import numpy as np

data = {
    'Name':   ['Rahul', 'Priya', 'Arjun', 'Sneha', 'Rahul', 'Meera', 'Vikram'],
    'Age':    [25, 28, None, 30, 25, 27, 999],
    'City':   ['Delhi ', 'Mumbai', 'delhi', 'Chennai', 'Delhi ', 'Delhi', None],
    'Marks':  [85, 92, 78, None, 85, 95, 65],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', None]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()           # Har cell True/False — null hai ya nahi
df.isnull().sum()     # Har column mein kitne nulls hain
df.isnull().sum() / len(df) * 100  # Percentage mein
```

```python
df.dropna()                        # Koi bhi null ho — poori row hata do
df.dropna(subset=['Age'])          # Sirf Age null ho toh row hata do
df.dropna(subset=['Age', 'Marks']) # Age ya Marks null ho toh hata do
df.dropna(thresh=4)                # Jo rows mein 4 se kam values hain hata do
```

```python
# Fixed value se fill karo
df['Marks'].fillna(0)

# Mean se fill karo — numeric columns ke liye best
df['Marks'].fillna(df['Marks'].mean())

# Median se fill karo — outliers ho toh median better hai
df['Age'].fillna(df['Age'].median())

# Mode se fill karo — categorical columns ke liye best
df['City'].fillna(df['City'].mode()[0])

# Forward fill — pichli value se fill karo
df['Marks'].fillna(method='ffill')

# Backward fill — agli value se fill karo
df['Marks'].fillna(method='bfill')



'''Permanently save karna na bhoolo!'''
# inplace=True se original df update hoga
df['Marks'].fillna(df['Marks'].mean(), inplace=True)

# Ya reassign karo
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# Duplicates check karo
df.duplicated()              # True/False — duplicate hai ya nahi
df.duplicated().sum()        # Kitne duplicates hain

# Duplicates dikhao
df[df.duplicated()]

# Duplicates hata do
df.drop_duplicates(inplace=True)

# Specific columns ke basis par duplicates hata do
df.drop_duplicates(subset=['Name', 'Age'], inplace=True)
```

```python
# Current dtypes check karo
df.dtypes

# String ko number mein convert karo
df['Age'] = df['Age'].astype(int)
df['Marks'] = df['Marks'].astype(float)

# Number ko string mein
df['Gender'] = df['Gender'].astype(str)

# String ko datetime mein — dates ke liye
df['Date'] = pd.to_datetime(df['Date'])
```

```python
'''string cleaning'''

# Extra spaces hatao — "Delhi " → "Delhi"
df['City'] = df['City'].str.strip()

# Sab uppercase karo
df['City'] = df['City'].str.upper()

# Sab lowercase karo
df['City'] = df['City'].str.lower()

# Pehla letter capital — "delhi" → "Delhi"
df['City'] = df['City'].str.title()

# Galat value replace karo
df['City'] = df['City'].str.replace('delhi', 'Delhi')
```

```python
'''Outliers handle karna'''

# Outlier check karo — describe se
df['Age'].describe()
# Age mein 999 clearly outlier hai

# Simple fix — condition se replace karo
df['Age'] = df['Age'].apply(lambda x: None if x > 100 else x)

# Phir fillna se handle karo
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
import pandas as pd
import numpy as np

data = {
    'Name':   ['Rahul', 'Priya', 'Arjun', 'Sneha', 'Vikram', 'Meera'],
    'Age':    [25, 28, 22, 30, 24, 27],
    'City':   ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Marks':  [85, 92, 78, 88, 65, 95],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'Salary': [30000, 55000, 25000, 62000, 28000, 70000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Simple calculation se
df['Marks_out_200'] = df['Marks'] * 2


# Do columns combine karo
df['Score_per_Age'] = df['Marks'] / df['Age']


# Condition se — Grade banana
df['Grade'] = df['Marks'].apply(lambda x:
    'A' if x >= 90 else
    'B' if x >= 80 else
    'C' if x >= 70 else 'D'
)


# Salary category
df['Salary_Category'] = df['Salary'].apply(lambda x:
    'High'   if x >= 60000 else
    'Medium' if x >= 40000 else 'Low'
)
df
```

```python
# Manual label encoding
df['Gender_Encoded'] = df['Gender'].map({'M': 0, 'F': 1})

# Ya pandas se
df['City_Encoded'] = df['City'].astype('category').cat.codes
```

```python
df
```

```python
# get_dummies — sabse easy tarika
city_dummies = pd.get_dummies(df['City'], prefix='City')
df = pd.concat([df, city_dummies], axis=1)
```

```python
# ============================================================
# Web Scraping kya hota hai?
# Website ka HTML padhna aur usme se data nikalna
# Jaise: product prices, news headlines, cricket scores
#
# Libraries:
# requests  — website ka HTML download karo
# BeautifulSoup — HTML mein se data nikalo
#
# Install karo pehle:
# pip install requests beautifulsoup4
# ============================================================

import requests
from bs4 import BeautifulSoup
import pandas as pd

# ============================================================
# EXAMPLE — Books scrape karenge
# Website: books.toscrape.com — practice ke liye bani hai
# ============================================================

# ----------------------------------------------------------
# Step 1 — Website ka HTML download karo
# ----------------------------------------------------------

url = "http://books.toscrape.com/"

# Browser jaisi request bhejo — warna site block kar sakti hai
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/91.0'
}

response = requests.get(url, headers=headers)

# Check karo sahi aaya ya nahi
print(f"Status Code: {response.status_code}")  # 200 = OK

# response.text mein poora HTML hai
print(response.text[:500])  # Pehle 500 characters dekho


# ----------------------------------------------------------
# Step 2 — BeautifulSoup se HTML parse karo
# Parse matlab — HTML ko samajhne layak structure mein todna
# ----------------------------------------------------------

soup = BeautifulSoup(response.text, 'html.parser')
# 'html.parser' — Python ka built-in HTML reader


# ----------------------------------------------------------
# Step 3 — HTML structure samjho
# Browser mein Right Click → Inspect karo
# Dekho data kaunse tag mein hai
# Yahan har book ek <article class="product_pod"> mein hai
# ----------------------------------------------------------

# Saari books dhundo
books = soup.find_all('article', class_='product_pod')
print(f"Kitni books mili: {len(books)}")  # 20 aani chahiye


# ----------------------------------------------------------
# Step 4 — Ek book se data nikalna samjho
# ----------------------------------------------------------

pehli_book = books[0]  # Pehli book lo

# Title nikalo — <h3> tag ke andar <a> tag mein hai
title = pehli_book.find('h3').find('a')['title']
print(f"Title: {title}")

# Price nikalo — <p class="price_color"> mein hai
price = pehli_book.find('p', class_='price_color').text
print(f"Price: {price}")

# Rating nikalo — <p class="star-rating One/Two/Three..."> mein hai
rating = pehli_book.find('p', class_='star-rating')['class'][1]
print(f"Rating: {rating}")

# Stock status — <p class="instock availability"> mein hai
stock = pehli_book.find('p', class_='instock availability').text.strip()
print(f"Stock: {stock}")


# ----------------------------------------------------------
# Step 5 — Saari books ka data nikalo — loop mein
# ----------------------------------------------------------

all_books = []  # Khali list — yahan data save karenge

for book in books:

    # Title
    title = book.find('h3').find('a')['title']

    # Price — £ sign hata ke float banao
    price = book.find('p', class_='price_color').text
    price = float(price.replace('£', '').replace('Â', '').strip())

    # Rating — words mein hai, numbers mein convert karo
    rating_word = book.find('p', class_='star-rating')['class'][1]
    rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
    rating = rating_map[rating_word]

    # Stock
    stock = book.find('p', class_='instock availability').text.strip()

    # Dictionary mein dalo
    all_books.append({
        'Title':  title,
        'Price':  price,
        'Rating': rating,
        'Stock':  stock
    })

print(f"\n✅ {len(all_books)} books ka data nikala!")


# ----------------------------------------------------------
# Step 6 — DataFrame banao aur analyse karo
# ----------------------------------------------------------

df = pd.DataFrame(all_books)
print(df.head())

# Ab saare purane methods use kar sakte ho!
print(df[df['Rating'] >= 4])              # 4+ rating wali books
print(df.sort_values('Price'))            # Price ke hisaab se sort
print(df['Price'].mean())                 # Average price
print(df.groupby('Rating')['Price'].mean()) # Rating wise avg price


# ----------------------------------------------------------
# Step 7 — Save karo
# ----------------------------------------------------------

df.to_csv('books_data.csv', index=False)
print("✅ books_data.csv save ho gaya!")


# ----------------------------------------------------------
# Step 8 — Multiple Pages scrape karna
# ----------------------------------------------------------

# Website mein page 1, page 2... hote hain
# URL pattern: books.toscrape.com/catalogue/page-2.html

all_pages_books = []

for page_num in range(1, 4):  # Page 1, 2, 3

    # Har page ka URL banao
    if page_num == 1:
        page_url = "http://books.toscrape.com/"
    else:
        page_url = f"http://books.toscrape.com/catalogue/page-{page_num}.html"

    print(f"Page {page_num} scrape ho raha hai...")

    # HTML download karo
    page_response = requests.get(page_url, headers=headers)
    page_soup = BeautifulSoup(page_response.text, 'html.parser')

    # Books dhundo
    page_books = page_soup.find_all('article', class_='product_pod')

    # Data nikalo
    for book in page_books:
        title  = book.find('h3').find('a')['title']
        price  = float(book.find('p', class_='price_color').text.replace('£','').replace('Â','').strip())
        rating = {'One':1,'Two':2,'Three':3,'Four':4,'Five':5}[book.find('p',class_='star-rating')['class'][1]]
        all_pages_books.append({'Title': title, 'Price': price, 'Rating': rating})

df_all = pd.DataFrame(all_pages_books)
print(f"\n✅ Total {len(df_all)} books scrape ho gayi!")
df_all.to_csv('all_books.csv', index=False)
```

```python
import pandas as pd
import numpy as np

data = {
    'Name':   ['Rahul', 'Priya', 'Arjun', 'Sneha', 'Vikram', 'Meera'],
    'Age':    [25, 28, 22, 30, 24, 27],
    'City':   ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Marks':  [85, 92, 78, 88, 65, 95],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['marks_to_grade']=df['Marks'].apply(lambda x:
    'A' if x >= 90 else
    'B' if x >= 80 else 'C'
)

df['Gender_Encoded'] = df['Gender'].map({'M': 0, 'F': 1})

city_dummies = pd.get_dummies(df['City'], drop_first=True)
df = pd.concat([df, city_dummies], axis=1)
df
```

```python
'''Marks column ko MinMaxScaler se 0 to 1 range mein scale karo'''

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Marks_Scaled'] = scaler.fit_transform(df[['Marks']])

print(df[['Name', 'Marks_Scaled']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ek achha dataset banate hain
data = {
    'Name':    ['Alice', 'Bob', 'Charlie', 'Diana', 'Evan', 'Fiona', 'George', 'Hannah'],
    'City':    ['Mumbai', 'Delhi', 'Mumbai', 'Chennai', 'Delhi', 'Bangalore', 'Mumbai', 'Chennai'],
    'Score':   [88, 75, 92, 68, 81, 95, 72, 85],
    'Age':     [25, 30, 22, 35, 28, 24, 31, 27],
    'Salary':  [50000, 65000, 48000, 72000, 58000, 80000, 55000, 67000]
}

df = pd.DataFrame(data)
df
```

```python
# Bar chart 

plt.figure(figsize=(8, 5))

city_counts = df['City'].value_counts()

plt.bar(city_counts.index, city_counts.values, color='steelblue')

plt.title('Har city mein kitne log hain')
plt.xlabel('City')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Line chart

plt.figure(figsize=(8, 5))

plt.plot(df['Name'], df['Score'], marker='o', color='green', linewidth=2)

plt.title('Har student ka Score')
plt.xlabel('Student')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##### Histogram

plt.figure(figsize=(8, 5))

plt.hist(df['Score'], bins=5, color='coral', edgecolor='black')

plt.title('Score distribution')
plt.xlabel('Score')
plt.ylabel('Kitne log')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot

plt.figure(figsize=(8, 5))

plt.scatter(df['Age'], df['Salary'], color='purple', s=100)

# Har point pe naam likhna
for i, name in enumerate(df['Name']):
    plt.annotate(name, (df['Age'][i], df['Salary'][i]), 
                 textcoords="offset points", xytext=(5, 5))

plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```
